```python
from graphviz import Digraph
```

```python
adj = {};

def addEdge(v, w):
    bag = adj.get(v, set())
    bag.add(w)
    adj[v] = bag

def removeEdge(v, w):
    bag = adj.get(v)
    if bag is not None:
        bag.remove(w)
        adj[v] = bag
```

```python
addEdge(0, 1)
addEdge(1, 0)
addEdge(1, 2)
addEdge(3, 4)
addEdge(0, 3)
addEdge(2, 0)
```

```python
print(adj)

#removeEdge(0,1)
#print(adj)
g = Digraph(comment = 'My graph')

for n in adj.keys():
    for a in adj.get(n):
        g.edge(str(n), str(a), str(n), color='red', arrowhead = 'open', style = 'dashed')

g
```
